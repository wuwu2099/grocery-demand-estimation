```python
ce_selected = ce_data.iloc[[3,4,5, 9,51], 0:7]
ce_selected.head()
```

```python
# set column names and row index 
ce = ce_selected.copy()

ce.columns = ce.iloc[0]
ce = ce.iloc[1:]
ce.columns.name = None

ce = ce.set_index(ce.columns[0])
ce.index.name = None

ce
```
